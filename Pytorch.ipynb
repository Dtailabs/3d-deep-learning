```python
#lets see how can we implement a transfer learning here
#step 1 import all the packages
import torch
from torch import nn, optim
import torch.nn.functional as F
import torchvision
from torchvision import transforms, datasets, models
import torch.autograd as grad
```

```python
#step 2 import all datasets and perform preprocessing on an input data
transform=transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5),(0.5))])
trainset=datasets.MNIST('~/.pytorch/MNIST',download=True, train=True, transform=transform)
testset=datasets.MNIST('~/.pytorch/MNIST',download=True, train=False, transform=transform)
Trainloader=torch.utils.data.DataLoader(trainset,batch_size=64,shuffle=True)
Testloader=torch.utils.data.DataLoader(testset,batch_size=64,shuffle=True)
```

```python
#step 3 import pretrained model and also write down a custom network
class Model(nn.Module):
  def __init__(self):
    super(Model,self).__init__()
    self.fc1=nn.Linear(784,512)
    self.fc2=nn.Linear(512,128)
    self.fc3=nn.Linear(128,10)
    self.dropout1=nn.Dropout(p=0.2)
    self.dropout2=nn.Dropout(p=0.2)
    self.logsoftmax=nn.LogSoftmax(dim=1)

  def forward(self,x):
    
    x=F.relu(self.dropout1(self.fc1(x)))
    x=F.relu(self.dropout2(self.fc2(x)))
    x=self.logsoftmax(self.fc3(x))
    return x
```

```python
model1=Model()
model1
```

```python
backbone=models.resnet50(pretrained=True)
for param in backbone.parameters():
  param.required_grad=False
```

```python
backbone.fc=model1
```

```python
backbone
```

```python
device=torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
for images, labels in Trainloader:
  print(images.shape,labels.shape)
```

```python
#Step 4 Now develop a pipeline to train a model
model1.to(device);
optimizer=optim.Adam(model1.parameters(),lr=0.001)
criterion=nn.NLLLoss()
epochs=10
for i in range(epochs):
  runningLoss=0
  for images, labels in Trainloader:
    images,labels=images.to(device),labels.to(device)
    images=images.view(images.shape[0],784)
    optimizer.zero_grad()
    outputs=model1.forward(images)
    loss=criterion(outputs,labels)
    loss.backward()
    optimizer.step()
    runningLoss+=loss.item()
    print('runningLoss:{}'.format(runningLoss/len(Trainloader)))
```

```python
with torch.no_grad():
  testLoss=0
  model1.eval()
  for images, labels in Testloader:
    images,labels=images.to(device),labels.to(device)
    images=images.view(images.shape[0],784)
    
    outputs=model1.forward(images)
    loss=criterion(outputs,labels)
    testLoss+=loss.item()
    print('runningLoss:{}'.format(testLoss/len(Testloader)))
```
